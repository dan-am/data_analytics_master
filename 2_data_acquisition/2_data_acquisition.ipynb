### Analyse der HR Analytics Job Change of Data Scientists Dataset

```python
import kagglehub
import pandas as pd
import os
import shutil

# Zielverzeichnis (bitte anpassen)
TARGET_DIR = r"/Users/danielmini/Projekte/data_analyics26/data_analytics_master/2_data_acquisition/raw_data"

# Dataset herunterladen (landet im kagglehub-Cache)
path = kagglehub.dataset_download(
    "arashnic/hr-analytics-job-change-of-data-scientists"
)
print("Download-Pfad:", path)
```

```python
# Zielverzeichnis anlegen
os.makedirs(TARGET_DIR, exist_ok=True)

# Dateien ins Zielverzeichnis kopieren
for file in os.listdir(path):
    src = os.path.join(path, file)
    dst = os.path.join(TARGET_DIR, file)
    if os.path.isfile(src):
        shutil.copy(src, dst)
```

```python

# CSV-Datei finden und laden
csv_files = [f for f in os.listdir(TARGET_DIR) if f.endswith(".csv")]
df = pd.read_csv(os.path.join(TARGET_DIR, csv_files[1]))

print("DataFrame geladen:", df.shape)
```
